## Data Preprocessing

```python
df_clean = df.copy()
```

```python
# menghapus baris yang memiliki nilai missing values pada kolom product_name dan customer_id
df_clean.dropna(subset=['product_name', 'customer_id'], inplace=True)
```

```python
#menghapus baris yang memiliki nilai negatif pada quantity dan price
df_clean = df_clean[(df_clean['quantity'] > 0) & (df_clean['price'] > 0)]
```

```python
#menghapus baris yang order idnya diawali dengna huruf c
df_clean = df_clean[~(df_clean['order_id'].str[:1] == 'C')]
```

```python
#menghapus data tester menggunakan product_code
df_clean = df_clean[~(df_clean['product_code'].str.lower().str.contains('test'))]
```

```python
#mengubah tipe data kolom customer_id dan order_date
df_clean['customer_id'] = df_clean['customer_id'].astype(str)
df_clean['order_date'] = pd.to_datetime(df_clean['order_date'])
```

```python
df_clean['amount'] = df_clean['quantity'] * df_clean['price']
```

```python
#mengganti product name menjadi seragam dari salah satu product name yang sering muncul
# grouping product_code dan product_name lalu menambahkan fungsi agregat, yakni nunique yang diberlakukan ke kolom order_id dan hasilnya dimasukkan ke new column, yakni order_cnt
# sorting nilai berdasarkan product code terlebih dahulu dari terkecil ke terbesar, baru diurutkan melalu order_cnt dari terbesar ke terkecil
most_freq_product_name = df_clean.groupby(['product_code', 'product_name'], as_index = False).agg(order_cnt=('order_id', 'nunique')).sort_values(['product_code','order_cnt'], ascending=[True,False])

#lalu buatkan rank dari product_name dalam satu grup(product_code) yang memiliki order_cnt terbanyak
most_freq_product_name['rank'] = most_freq_product_name.groupby('product_code')['order_cnt'].rank(method='first', ascending=False)

#filter isi most_freq_product_name hingga hanya terdiri dari product_name yang memiliki order_cnt terbanyak
most_freq_product_name = most_freq_product_name[most_freq_product_name['rank'] == 1].drop(columns=['order_cnt', 'rank'])

#lakukan merge pada df_clean dengan most_freq_product_name dimana kolom product_name pada most_freq_product_name di rename terlebih dahulu
#lakukan merge secara left di mana menyesuaikan dengan product code di df_clean
df_clean = df_clean.merge(most_freq_product_name.rename(columns={'product_name':'most_freq_product_name'}), how='left', on='product_code')

#ganti nama produk dengan most_freq_product_name
df_clean['product_name'] = df_clean['most_freq_product_name'].str.lower()

df_clean = df_clean.drop(columns='most_freq_product_name')
```

```python
#menghapus outlier
df_clean = df_clean[(np.abs(stats.zscore(df_clean[['quantity','amount']]))<3).all(axis=1)]
df_clean = df_clean.reset_index(drop=True)
df_clean
```

## Menyiapkan data basket

```python

basket = pd.pivot_table(df_clean,index='order_id', columns='product_name', values='product_code', aggfunc='nunique', fill_value=0)
basket
```

```python
def encode(x):
  if x == 0:
    return False
  else :
    return True

basket_encode = basket.applymap(encode)
basket_encode
```

```python
#filter sehingga minimal terdapat 2 barang yang terbeli dalam suatu order
basket_filter = basket_encode[(basket_encode>0).sum(axis=1)>1]
basket_filter
```

```python
basket_filter.info()
```

# Implementing Algorithm

```python
frequent_itemset = apriori(basket_filter, min_support=0.01, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemset['product_cnt']=frequent_itemset['itemsets'].apply(lambda x : len(x))
frequent_itemset
```

```python
#hitung confidence, lift, support
product_association = association_rules(frequent_itemset,
                                        metric='confidence',
                                        min_threshold=0.7,
                                        num_itemsets=None).sort_values(['support', 'confidence'], ascending=[False, False]).reset_index(drop=True)

product_association
```
